```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib.pylab import rcParams
rcParams['figure.figsize']=20,10
from keras.models import Sequential
from keras.layers import LSTM,Dropout,Dense

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
```

```python
df_nse = pd.read_csv("HDFC.csv")

df_nse["Date"]=pd.to_datetime(df_nse.Date,format="%d-%m-%Y")
df_nse.index=df_nse['Date']


data=df_nse.sort_index(ascending=True,axis=0)
new_data1=pd.DataFrame(index=range(0,len(df_nse)),columns=['Date','Close'])
new_data_date1=pd.DataFrame(index=range(0,len(df_nse)),columns=['Date',])

for i in range(0,len(data)):
    new_data1["Date"][i]=data['Date'][i]
    new_data1["Close"][i]=data["Close"][i]
    new_data_date1["Date"][i]=data['Date'][i]
#new_data.index=new_data.Date
new_data1.drop("Date",axis=1,inplace=True)

dataset=new_data1.values
test_date1 = new_data_date1["Date"][4160:]
train1=dataset[0:4160,:]
valid1=dataset[4160:,:]

scaler=MinMaxScaler(feature_range=(0,1))
scaled_data=scaler.fit_transform(dataset)

x_train1,y_train1=[],[]

for i in range(60,len(train1)):
    x_train1.append(scaled_data[i-60:i,0])
    y_train1.append(scaled_data[i,0])
    
x_train1,y_train1=np.array(x_train1),np.array(y_train1)

x_train1=np.reshape(x_train1,(x_train1.shape[0],x_train1.shape[1],1))
```

```python
lstm_model=Sequential()
lstm_model.add(LSTM(units=50,return_sequences=True,input_shape=(x_train1.shape[1],1)))
lstm_model.add(LSTM(units=50))
lstm_model.add(Dense(1))
inputs_data=new_data1[len(new_data1)-len(valid1)-60:].values
inputs_data=inputs_data.reshape(-1,1)
inputs_data=scaler.transform(inputs_data)
lstm_model.compile(loss='mean_squared_error',optimizer='adam')
lstm_model.fit(x_train1,y_train1,epochs=5,batch_size=1,verbose=2)
```

```python
lstm_model.save("saved_model1.h5")

lstm_model=load_model("saved_model1.h5")
inputs_data=new_data1[len(new_data1)-len(valid1)-60:].values
inputs_data=inputs_data.reshape(-1,1)
inputs_data=scaler.transform(inputs_data)
X_test1=[]
for i in range(60,inputs_data.shape[0]):
    X_test1.append(inputs_data[i-60:i,0])
X_test1=np.array(X_test1)
X_test1=np.reshape(X_test1,(X_test1.shape[0],X_test1.shape[1],1))
predicted_closing_price1=lstm_model.predict(X_test1)
predicted_closing_price1=scaler.inverse_transform(predicted_closing_price1)
stock_pred1=new_data1[4160:]
stock_pred1['predictions']=predicted_closing_price1
```
